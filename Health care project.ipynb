## Step 2: Data Cleaning & Manipulation

## 1. Handle Missing Data

```python
df.isnull().sum()
```

## Observation : 
         ->No missing values (55500 non-null for each column)
         ->Missing values reduce model performance and distort statistical analysis.

## 2. Remove Duplicates

```python
df.duplicated().sum()
```

## Observation : 
          It will Check for duplicates

```python
df = df.drop_duplicates()
df
```

## Observation :
          -> If duplicates exist drop will remove them
          -> Duplicates can inflate counts (e.g., number of admissions), misleading insights.

## 3. Fix Inconsistencies in Categorical Data

```python
df['Medical Condition'] = df['Medical Condition'].str.strip().str.title()
df
```

## Observation :
           -> Consistent categories prevent fragmentation (e.g., “urgent” vs “Urgent” counted as separate groups).

## 4. Convert Data Types

## Date Columns :
             -> Convert Date of Admission & Discharge Date:

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df
```

```python
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df
```

## Categorical Columns

```python
    -> Convert large repetitive text columns to category type
```

```python
cat_cols = ['Gender','Blood Type','Medical Condition','Doctor',
            'Hospital','Insurance Provider','Admission Type',
            'Medication','Test Results']

for col in cat_cols:
    df[col] = df[col].astype('category')
df    
```

## Observation :
         -> Reduces memory footprint
         -> Enables faster computations
         -> Ensures proper data handling
